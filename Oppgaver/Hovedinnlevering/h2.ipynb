Dersom kretsen ikke finnes returneres 'ukjent krets'

```python
def lesKretser():
    kretser = {}
    with open('kretser.txt', 'r') as fil:
        for linje in fil:
            krets_nummer, krets_verdi = linje.split(',')
            kretser[krets_nummer] = krets_verdi.strip()
    return kretser

# NYTT *******************
# Definerer en funksjon som tar en krets som argument og returnerer kretsnummeret
def kretsNr(krets):
    # Konverterer krets til en streng for å sikre sammenlignbarhet
    krets = str(krets)

    # Itererer gjennom ordboken med kretser
    for krets_kode, krets_verdi in kretser.items():
        # Sjekker om krets matcher enten kretsnummeret eller kretsverdien
        if krets == krets_kode or krets == krets_verdi:
            # Returnerer kretsnummeret som en integer hvis det er en match
            return int(krets_kode)
        
    # Returnerer 'ukjent krets' hvis der ikke ble funnet noen match
    return 'ukjent krets'
# ************************

kretser = lesKretser()

# EKSEMPEL *******************
print(kretsNr('Damsgård'))
print(kretsNr('Sentrum'))
print(kretsNr(74))
```

```python
class StemmeTall:
    def __init__(self, parti_kode, stemmer):
        self.parti_kode = parti_kode
        self.stemmer = stemmer

    def skriv(self):
        print(f"{self.parti_kode}:{self.stemmer}")

def lesKretser():
    kretser = {}
    with open('kretser.txt', 'r') as fil:
        for linje in fil:
            krets_nummer, krets_verdi = linje.split(',')
            kretser[krets_nummer] = krets_verdi.strip()
    return kretser

def kretsNr(krets):
    krets = str(krets)

    for krets_kode, krets_verdi in kretser.items():
        if krets == krets_kode or krets == krets_verdi:
            return krets_kode
        
    return 'ukjent krets'

# NYTT *******************
# Definerer en funksjon som leser valgresultater fra en fil og returnerer en ordbok med kretsnummer som nøkkel og en liste av StemmeTall-objekter som verdi
def lesValg(filnavn):
    # Initialiserer en tom ordbok for resultatene
    resultat = {}
    # Åpner filen i 'lesemodus'
    with open(filnavn, 'r') as fil:
        # Leser og ignorerer den første linjen (overskriften)
        fil.readline()
        # Itererer gjennom hver linje i filen
        for linje in fil:
            # Deler linjen ved komma for å få kretsnummer, partiets kode og antall stemmer
            krets_nr, parti_kode, stemmer = linje.split(',')
            # Sjekker om kretsnummeret allerede finnes i resultat-ordboken
            if krets_nr not in resultat:
                # Hvis ikke, oppretter en ny liste for kretsnummeret
                resultat[krets_nr] = []
            # Legger til et StemmeTall-objekt med partiets kode og antall stemmer til listen for kretsnummeret
            resultat[krets_nr].append(StemmeTall(parti_kode, int(stemmer.strip())))
    # Returnerer ordboken med valgresultatene
    return resultat
# ************************

# EKSEMPEL *******************
kretser = lesKretser()
valg13=lesValg('stemmer2013.txt')
Damsgård=kretsNr('Damsgård')
for stemmetall in valg13[Damsgård]: stemmetall.skriv()
```

```python
class Parti:
    def __init__(self, parti_kode, parti_navn):
        self.parti_kode = parti_kode
        self.parti_navn = parti_navn
        self.parti_valgår = []
    
    def skriv(self):
        print(f"{self.parti_navn} ({self.parti_kode})", end=" ")
        print(f"Valgår: {', '.join(set(self.parti_valgår))}" if self.parti_valgår else "")

class StemmeTall:
    def __init__(self, parti_kode, stemmer):
        self.parti = parti_kode
        self.antall = stemmer

    def skriv(self):
        print(f"{self.parti}:{self.antall}")

def lesValg(filnavn, valgår):
    resultat = {}

    with open(filnavn, 'r') as fil:
        fil.readline()
        for linje in fil:
            krets_nr, parti_kode, stemmer = linje.split(',')

            if krets_nr not in resultat:
                resultat[krets_nr] = []

            resultat[krets_nr].append(StemmeTall(parti_kode, int(stemmer.strip())))
            for p in partier:
                if p.parti_kode == parti_kode and str(valgår) not in p.parti_valgår:
                    p.parti_valgår.append(str(valgår))
    return resultat

def lesPartier():
    partier = []
    with open('partier.txt', 'r') as fil:
        for linje in fil:
            kode, navn = linje.split(',')
            partier.append(Parti(kode.strip(), navn.strip()))

    return partier

def samlet(valg):
    samlet_valg = {}

    for partier in valg.values():
        for p in partier:
            if p.parti not in samlet_valg:
                samlet_valg[p.parti] = 0
            samlet_valg[p.parti] += p.antall

    return samlet_valg

# NYTT *******************
# Definerer en funksjon som beregner prosentfordelingen av stemmer for hvert parti
def prosentfordeling(valg):
    # Beregner det totale antallet stemmer ved å summere verdiene i den samlede valgresultat-ordboken
    total_stemmer = sum(samlet(valg).values())
    # Returnerer en ordbok der nøkkelen er partiets kode og verdien er prosentandelen av stemmene
    # Prosentandelen beregnes ved å dele antall stemmer for hvert parti på total_stemmer og multiplisere med 100
    # Resultatet rundes av til to desimaler
    return {parti: round(stemmer / total_stemmer * 100, 2) for parti, stemmer in samlet(valg).items()}
# ************************

partier = lesPartier()

valg13=lesValg('stemmer2013.txt',2013)
valg17=lesValg('stemmer2017.txt',2017)
valg21=lesValg('stemmer2021.txt',2021)

# EKSEMPEL *******************
prosentfordeling(valg21)
```

```python
class Parti:
    def __init__(self, parti_kode, parti_navn):
        self.parti_kode = parti_kode
        self.parti_navn = parti_navn
        self.parti_valgår = []
    
    def skriv(self):
        print(f"{self.parti_navn} ({self.parti_kode})", end=" ")
        print(f"Valgår: {', '.join(set(self.parti_valgår))}" if self.parti_valgår else "")

class StemmeTall:
    def __init__(self, parti_kode, stemmer):
        self.parti = parti_kode
        self.antall = stemmer

    def skriv(self):
        print(f"{self.parti}:{self.antall}")

def kretsNr(krets):
    krets = str(krets)

    for krets_kode, krets_verdi in kretser.items():
        if krets == krets_kode or krets == krets_verdi:
            return krets_kode
        
    return 'ukjent krets'

def kretsNavn(krets):
    krets = str(krets)

    for krets_kode, krets_verdi in kretser.items():
        if krets == krets_kode or krets == krets_verdi:
            return krets_verdi
        
    return 'ukjent krets'

def partiKode(parti_navn_søk):
    for parti_klasse in partier:
        if parti_klasse.parti_navn == parti_navn_søk or parti_klasse.parti_kode == parti_navn_søk:
            return parti_klasse.parti_kode
    return 'ukjent parti'

def lesValg(filnavn, valgår):
    resultat = {}

    with open(filnavn, 'r') as fil:
        fil.readline()
        for linje in fil:
            krets_nr, parti_kode, stemmer = linje.split(',')

            if krets_nr not in resultat:
                resultat[krets_nr] = []

            resultat[krets_nr].append(StemmeTall(parti_kode, int(stemmer.strip())))
            for p in partier:
                if p.parti_kode == parti_kode and str(valgår) not in p.parti_valgår:
                    p.parti_valgår.append(str(valgår))
    return resultat

def lesPartier():
    partier = []
    with open('partier.txt', 'r') as fil:
        for linje in fil:
            kode, navn = linje.split(',')
            partier.append(Parti(kode.strip(), navn.strip()))

    return partier

def lesKretser():
    kretser = {}
    with open('kretser.txt', 'r') as fil:
        for linje in fil:
            krets_nummer, krets_verdi = linje.split(',')
            kretser[krets_nummer] = krets_verdi.strip()
    return kretser

def samlet(valg):
    samlet_valg = {}

    for partier in valg.values():
        for p in partier:
            if p.parti not in samlet_valg:
                samlet_valg[p.parti] = 0
            samlet_valg[p.parti] += p.antall

    return samlet_valg

def prosentfordeling(valg):
    total_stemmer = sum(samlet(valg).values())
    return {parti: round(stemmer / total_stemmer * 100, 2) for parti, stemmer in samlet(valg).items()}

# NYTT *******************
# Definerer en funksjon som gir en oversikt over valgresultatet for et gitt parti
def partiOversikt(valg, parti):
    # Finner partiets kode basert på partiets navn eller kode
    parti_kode = partiKode(parti)
    # Initialiserer en tom ordbok for å lagre antall stemmer for partiet i hver krets
    parti_stemmer = {}
    # Initialiserer en tom ordbok for å lagre det totale antallet stemmer i hver krets
    total_stemmer_krets = {}

    # Itererer gjennom alle kretser og partilister i valgresultatene
    for krets, partier in valg.items(): 
        # Beregner det totale antallet stemmer i kretsen 
        antall_stemmer_krets = sum(p.antall for p in partier)
        # Lagrer det totale antallet stemmer i kretsen i ordboken
        total_stemmer_krets[krets] = antall_stemmer_krets

        # Itererer gjennom hver parti i partilisten
        for p in partier:
            # Sjekker om partiets kode matcher den søkte partikoden
            if p.parti == parti_kode:
                # Lagrer antall stemmer for partiet i kretsen i ordboken
                parti_stemmer[krets] = p.antall

    # Beregner prosentfordelingen av stemmer for partiet i hver krets
    prosent_fordeling_stemmer = {
        krets: (stemmer / total_stemmer_krets[krets]) * 100
        for krets, stemmer in parti_stemmer.items()
    }

    # Finner den høyeste prosentandelen av stemmer for partiet
    høyeste_oppsluttning = round(max(prosent_fordeling_stemmer.values()), 2)
    # Finner den laveste prosentandelen av stemmer for partiet
    laveste_oppsluttning = round(min(prosent_fordeling_stemmer.values()), 2)
    
    # Finner kretsene med den høyeste prosentandelen av stemmer for partiet
    høyest_oppsluttning_krets = [
        kretsNavn(krets) for krets, oppsluttning in prosent_fordeling_stemmer.items()
        if round(oppsluttning, 2) == høyeste_oppsluttning
    ]
    
    # Finner kretsene med den laveste prosentandelen av stemmer for partiet
    lavest_oppsluttning_krets = [
        kretsNavn(krets) for krets, oppsluttning in prosent_fordeling_stemmer.items()
        if round(oppsluttning, 2) == laveste_oppsluttning
    ]

    # Skriver ut det samlede antallet stemmer og prosentandelen av stemmene for partiet
    print(f"{parti} fikk {samlet(valg)[parti_kode]} stemmer ({prosentfordeling(valg)[parti_kode]}%)")
    # Skriver ut kretsene med den høyeste prosentandelen av stemmer for partiet
    print(f"Høyest oppsluttning ({høyeste_oppsluttning})%: {', '.join(høyest_oppsluttning_krets)}")
    # Skriver ut kretsene med den laveste prosentandelen av stemmer for partiet
    print(f"Lavest oppsluttning ({laveste_oppsluttning})%: {', '.join(lavest_oppsluttning_krets)}")
# ************************

kretser = lesKretser()
partier = lesPartier()
valg13=lesValg('stemmer2013.txt',2013)
valg17=lesValg('stemmer2017.txt',2017)
valg21=lesValg('stemmer2021.txt',2021)

# EKSEMPEL *******************
partiOversikt(valg21, 'Industri- og Næringspartiet')
```
